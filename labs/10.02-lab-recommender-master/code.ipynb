```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
desc = pd.read_csv('./product_descriptions.csv')
desc.head()
```

```python
desc.shape
```

```python
data = desc.sample(frac=.15)
```

```python
cvect = CountVectorizer()
sparse = cvect.fit_transform(data['product_description'])
sparse
```

```python
similarity_matrix = cosine_similarity(sparse)
```

```python
recommender_df = pd.DataFrame(similarity_matrix, index=data['product_description'], columns=data['product_description'])
recommender_df.head(2)
```

```python
recommender_df.iloc[:10,5].sort_values(ascending=False)
```

# Interestingly the recommender does poorly when matching products, but the adjectives and general style of the descriptions are well matched
