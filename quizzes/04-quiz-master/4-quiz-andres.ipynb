```python
from requests import get
import pandas as pd
```

From reading the API docs I'm going to select the following features for every pokemon:
- id
- base_experience (TARGET)
- height
- is_default
- weight

```python
# Perform first pull to check data
url = 'https://pokeapi.co/api/v2/pokemon/1/'
res = get(url)
res.json()['base_experience']
```

**BEWARE! THIS NEXT CELL TAKES SEVERAL MINUTES TU RUN**

```python
# With help from BreakfastHour code
# Outputs collapsed
url_root = 'https://pokeapi.co/api/v2/pokemon/'

poke_num = 1
error_count = 0 
pokemon= []

while error_count < 5:
    url = url_root + str(poke_num)
    res = get(url)
    if res.status_code == 200:
        error_count = 0
        data = res.json()
        columns = ['id', 'base_experience', 'height',  'is_default', 'weight']
        poke = {}
        for col in columns:
            poke[col] = data[col]
        pokemon.append(poke)
        poke_num += 1
        print(f'{data["name"]} was added!')
    else:
        error_count += 1 
        print(f'Pokemon number {poke_num} does not exist!')
        poke_num += 1

pd.DataFrame(pokemon)
```

```python
# Rename target column
poke_df = pd.DataFrame(pokemon)
poke_df.rename(columns={'base_experience': 'base_exp_TARGET'})
```
